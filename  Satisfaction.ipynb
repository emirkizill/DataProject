```python
##Libraries
import numpy as np
import seaborn as sns 
import pandas as pd
```

```python
##Add Dataset
sales=pd.read_csv('supermarket_sales.csv')
```

```python
ds=sales.copy()
```

## Which gender is more satisfied?

```python
ds.groupby(["Gender"])["Rating"].mean()
```

## Gender Score Distribution

```python
sns.kdeplot(x= "Rating",hue = "Gender",data=ds)
```

## Which city is more satisfied?

```python
ds.groupby(["City"])["Rating"].mean()
```

```python
sns.barplot(x="City",y="Rating",hue = "Gender",data=ds)
```

## Which product are customers more satisfied with?

```python
ds.groupby(["Product line"])["Rating"].mean()
```

```python
sns.barplot(x="City",y="Rating",hue="Product line",data = ds);
```

## Which gender was more satisfied with which product?

```python
sns.barplot(x="Gender",y="Rating",hue="Product line",data=ds)
```

## Satisfaction distribution according to the price paid

```python
sns.lmplot(x="Total",y="Rating",hue = "Gender",col = "City",data = ds);
```
