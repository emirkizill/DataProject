```python
##Libraries
import numpy as np
import seaborn as sns 
import pandas as pd
```

```python
##Add Dataset
sales=pd.read_csv('supermarket_sales.csv')
```

```python
ds=sales.copy()
```

## Which gender is more satisfied?

```python
ds.groupby(["Gender"])["Rating"].mean()
```

```python
sns.kdeplot(x= "Rating",hue = "Gender",data=ds)
```
